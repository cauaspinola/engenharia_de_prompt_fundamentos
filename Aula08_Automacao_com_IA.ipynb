```python
# Script para verificar a temperatura do sistema

try:
    # Início de um bloco try-except para lidar com possíveis erros de entrada.
    # Simulando a obtenção da temperatura. Em um sistema real, você obteria isso de um sensor ou API.
    temperatura_str = input('Digite a temperatura atual do sistema em Celsius: ') # Solicita a temperatura ao usuário como string
    temperatura = float(temperatura_str) # Converte a string de temperatura para um número decimal (float)

    # Verifica se a temperatura é maior que 30 graus Celsius
    if temperatura > 30:
        # Se for maior, exibe um alerta de superaquecimento
        print('ALERTA: Superaquecimento! Temperatura atual: ' + str(temperatura) + '°C. Recomenda-se verificar o sistema.')
    else:
        # Se não for maior que 30, a temperatura é considerada normal
        print('Temperatura do sistema: ' + str(temperatura) + '°C. Tudo parece normal.')
except ValueError:
    # Se ocorrer um erro ao converter a entrada para float (ex: usuário digitou texto), exibe uma mensagem de erro
    print('Erro: Por favor, digite um número válido para a temperatura.')
```

```python
import os
import shutil
import random
import string

# 1. Definir o diretório de teste
# O diretório onde os arquivos de exemplo serão criados e organizados.
# Se já existir, garantimos que ele esteja limpo para o teste.
test_directory = "arquivos_baguncados"

# Tenta remover o diretório e seu conteúdo se já existir, para um teste limpo.
if os.path.exists(test_directory):
    shutil.rmtree(test_directory)
# Cria o diretório de teste.
os.makedirs(test_directory)

print("Diretório de teste '{}' criado.".format(test_directory))

# 2. Criar 10 arquivos de exemplo
# Lista de nomes de arquivos para serem criados.
example_files = [
    "documento_importante.pdf",
    "foto_ferias.jpg",
    "relatorio_final.docx",
    "musica_favorita.mp3",
    "video_engracado.mp4",
    "nota_simples.txt",
    "apresentacao.pptx",
    "planilha_dados.xlsx",
    "backup.zip",
    "script_python.py",
    "outra_foto.png",
    "arquivo_log.log"
]

# Cria os arquivos vazios no diretório de teste.
for filename in example_files:
    file_path = os.path.join(test_directory, filename)
    with open(file_path, "w") as f:
        f.write("Conteúdo de exemplo para {}".format(filename))
    print("Arquivo '{}' criado.".format(filename))

print("\n--- Iniciando a organização dos arquivos ---")

# Dicionário para mapear extensões a nomes de pastas mais descritivos (opcional)
# Caso não esteja aqui, a pasta será criada com o nome da extensão (ex: "pdf", "jpg")
folder_names = {
    ".pdf": "PDFs",
    ".jpg": "Imagens",
    ".jpeg": "Imagens",
    ".png": "Imagens",
    ".gif": "Imagens",
    ".txt": "Textos",
    ".docx": "Documentos",
    ".xlsx": "Planilhas",
    ".pptx": "Apresentações",
    ".mp3": "Músicas",
    ".mp4": "Vídeos",
    ".avi": "Vídeos",
    ".mov": "Vídeos",
    ".zip": "Compactados",
    ".rar": "Compactados",
    ".py": "Scripts Python",
    ".log": "Logs"
}

# Percorrer todos os itens no diretório
for item in os.listdir(test_directory):
    # Construir o caminho completo para o item
    item_path = os.path.join(test_directory, item)

    # Verificar se é um arquivo (e não uma subpasta que já pode ter sido criada)
    if os.path.isfile(item_path):
        # Separar o nome do arquivo da extensão
        _, extension = os.path.splitext(item)
        # Converter a extensão para minúsculas para padronização
        extension = extension.lower()

        # Determinar o nome da subpasta
        # Se a extensão estiver no dicionário, usa o nome descritivo; senão, usa a própria extensão (sem o ponto)
        if extension in folder_names:
            subfolder_name = folder_names[extension]
        else:
            subfolder_name = extension[1:] + " (Outros)" # Mantendo a concatenação simples

        # Construir o caminho completo para a subpasta
        subfolder_path = os.path.join(test_directory, subfolder_name)

        # Criar a subpasta se ela não existir
        if not os.path.exists(subfolder_path):
            os.makedirs(subfolder_path)
            print("Pasta '{}' criada.".format(subfolder_name))

        # Construir o caminho de destino do arquivo
        destination_path = os.path.join(subfolder_path, item)

        # Mover o arquivo para a subpasta
        shutil.move(item_path, destination_path)
        print("Arquivo '{}' movido para '{}'.".format(item, subfolder_name))

print("\nOrganização concluída!")
print("Verifique o diretório '{}' para ver os resultados.".format(test_directory))
```

```python
import requests
import json

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço de um determinado CEP.
    Retorna um dicionário com os dados do endereço ou None se o CEP for inválido/não encontrado.
    """
    # Constrói a URL para a consulta da API ViaCEP usando concatenação de strings
    url = "https://viacep.com.br/ws/" + cep + "/json/"
    try:
        # Realiza a requisição GET para a API
        response = requests.get(url)
        # Lança um erro para status HTTP ruins (4xx ou 5xx)
        response.raise_for_status()
        # Decodifica a resposta JSON
        dados = response.json()

        # Verifica se a API retornou um erro (CEP não encontrado/inválido)
        if 'erro' in dados and dados['erro']:
            # Imprime mensagem de erro usando concatenação
            print("CEP '" + cep + "' não encontrado ou inválido.")
            return None
        else:
            # Retorna os dados do endereço
            return dados
    # Captura exceções relacionadas a problemas na requisição (ex: rede, URL inválida)
    except requests.exceptions.RequestException as e:
        # Imprime mensagem de erro de requisição usando concatenação
        print("Erro ao consultar o CEP " + cep + ": " + str(e))
        return None
    # Captura exceções relacionadas à decodificação JSON
    except json.JSONDecodeError:
        # Imprime mensagem de erro de decodificação JSON usando concatenação
        print("Erro ao decodificar a resposta JSON para o CEP " + cep + ".")
        return None

# Exemplo de uso:
# Solicita ao usuário que digite um CEP para consulta
cep_exemplo = input("Digite um CEP para consultar (ex: 01001000): ")

# Verifica se o usuário digitou algo
if cep_exemplo:
    # Remove qualquer caractere não numérico do CEP para garantir que seja apenas dígitos
    cep_limpo = ''.join(filter(str.isdigit, cep_exemplo))

    # Verifica se o CEP limpo tem 8 dígitos
    if len(cep_limpo) == 8:
        # Chama a função para consultar o CEP
        endereco = consultar_cep(cep_limpo)

        # Se o endereço foi encontrado, imprime as informações formatadas
        if endereco:
            print("\n--- Informações do Endereço ---")
            # Imprime o dicionário de endereço formatado como JSON, com indentação para legibilidade
            # e ensure_ascii=False para exibir caracteres especiais corretamente
            print(json.dumps(endereco, indent=4, ensure_ascii=False))
        else:
            # Mensagem caso não seja possível obter o endereço (já tratada na função consultar_cep)
            pass
    else:
        # Mensagem para CEP com número de dígitos incorreto
        print("Por favor, digite um CEP válido com 8 dígitos.")
else:
    # Mensagem caso nenhum CEP tenha sido digitado
    print("Nenhum CEP foi digitado.")
```

```python
import time
import random

def check_critical_condition():
    # Esta função simula a verificação de uma condição crítica.
    # Em um cenário real, você buscaria dados de sensores, APIs, bancos de dados, etc.
    # Aqui, vamos simular uma "temperatura" aleatória.
    current_value = random.uniform(15.0, 45.0) # Valor entre 15 e 45 para simulação
    critical_threshold = 35.0 # Define o limite crítico

    print(f"Verificando... Valor atual: {current_value:.2f}°C")

    if current_value > critical_threshold:
        return True, f"ALERTA CRÍTICO: Valor ({current_value:.2f}°C) excedeu o limite de {critical_threshold}°C!"
    else:
        return False, f"Tudo normal. Valor atual: {current_value:.2f}°C."

def send_alert(message, to_email="admin@example.com"):
    # Esta função simula o envio de um alerta.
    # Em um cenário real, você usaria um módulo como 'smtplib' para enviar e-mails reais,
    # ou integraria com um serviço de mensagens/notificações.
    print("\n" + "="*40)
    print("!!! ALERTA DISPARADO !!!")
    print(f"Mensagem para o console: {message}")
    print(f"Simulando envio de e-mail para {to_email}: Assunto: ALERTA CRÍTICO! Mensagem: {message}")
    print("="*40 + "\n")

def autonomous_program(interval_seconds=5, max_checks=10):
    print("Iniciando programa autônomo de monitoramento.")
    checks_done = 0
    while checks_done < max_checks:
        is_critical, message = check_critical_condition()
        if is_critical:
            send_alert(message)

        checks_done += 1
        print(f"Próxima verificação em {interval_seconds} segundos... ({checks_done}/{max_checks})")
        time.sleep(interval_seconds)
    print("Programa autônomo finalizado.")

# Para executar o programa autônomo:
# O programa fará 10 verificações, com 5 segundos de intervalo entre cada uma.
# Você pode ajustar 'interval_seconds' e 'max_checks' conforme necessário.
autonomous_program(interval_seconds=3, max_checks=5)
```
